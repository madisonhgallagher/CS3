```python
# First, install the required packages if you haven't already
!pip install matplotlib pandas

# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Create a data frame with the genre statistics
data = {
    'City': ['Los Angeles', 'New York City', 'Chicago', 'Houston', 'Miami', 'Washington, D.C.', 'Atlanta'],
    'Hip-Hop/Rap': [40, 24, 40, 32, 28, 32, 48],
    'R&B/Soul': [24, 24, 24, 24, 16, 24, 16],
    'Latin': [20, 20, 20, 24, 24, 20, 0],
    'Country': [4, 4, 4, 8, 8, 8, 12]
}

df = pd.DataFrame(data)

# Set the City column as the index
df.set_index('City', inplace=True)

# Plot the data
ax = df.plot(kind='bar', stacked=False, figsize=(12, 8), colormap='tab10')
ax.set_title('Genre Distribution Across Cities', fontsize=15)
ax.set_xlabel('City', fontsize=12)
ax.set_ylabel('Percentage', fontsize=12)
ax.legend(title='Genre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a data frame with the top three artists' song counts
data_artists = {
    'City': ['Los Angeles', 'New York City', 'Chicago', 'Houston', 'Miami', 'Washington, D.C.', 'Atlanta'],
    'Kendrick Lamar': [5, 3, 4, 3, 3, 3, 0],
    'The Weeknd': [3, 3, 3, 3, 3, 3, 0],
    'Bad Bunny': [2, 4, 3, 3, 4, 2, 0],
    'Lil Baby': [0, 0, 0, 0, 0, 0, 7],
    'Morgan Wallen': [0, 0, 0, 0, 3, 3, 3],
    'Rod Wave': [0, 0, 0, 0, 0, 0, 2]
}

df_artists = pd.DataFrame(data_artists)

# Set the City column as the index
df_artists.set_index('City', inplace=True)

# Plot the data for top three artists' song counts
ax_artists = df_artists.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab20')
ax_artists.set_title('Top Three Artists\' Song Counts Across Cities', fontsize=15)
ax_artists.set_xlabel('City', fontsize=12)
ax_artists.set_ylabel('Number of Songs', fontsize=12)
ax_artists.legend(title='Artist')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
